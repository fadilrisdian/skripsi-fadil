```python
plants = pd.read_csv('dataset/interim/results_plants_completed.csv')
```

```python
driver = webdriver.Chrome()
driver.get("http://www.knapsackfamily.com/KNApSAcK_World/search.php?cn=IDN&wd=&flg=")
html=driver.page_source
soup=BeautifulSoup(html,'html.parser')
soup
```

```python
# # spesies error ketika diproses ke regex
plants = plants.drop(labels=[2411], axis=0)
plants = plants.drop(labels=[2871], axis=0)
# #PR -> bikin code ketika kode error bisa lanjut looping
```

```python
i = 1
flag_array = []
for spesies in plants['Species']:
    print('Spesies', spesies)
    if(len(soup(text=re.compile(spesies)))>0):
        print('Found')
        flag_array.append("Yes")
    else:
        print('Not Found')
        flag_array.append(None)
    print('plant ke ' + str(i))
    i += 1
```

```python
flag_array_np = np.array(flag_array)
plants['Indonesia available?'] = flag_array_np
plants.to_csv('dataset/interim/Plant_Indo_Raw.csv',index = False)
```
