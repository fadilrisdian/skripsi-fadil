## Get Senyawa based on Name

```python
data_herbal= pd.read_csv('D:\Thesis\Data\HasilPredictSenyawaHerbalDaylight40k.csv')
data_herbal
```

```python
driver = webdriver.Chrome('D:/chromedriver')
plants_df=pd.DataFrame()
list_plants = []
k=0
l=0
m=0
for compounds in data_herbal['Senyawa']:
    try:
        driver.switch_to.window(driver.window_handles[0])
        driver.get("http://www.knapsackfamily.com/knapsack_core/result.php?sname=all&word="+str(compounds))
        print("Mencari tanaman senyawa:", compounds)
        #klik cid senyawa
        time.sleep(7)
        target = driver.find_element_by_xpath("//td[@class='d1'][1]/a")
        actions = ActionChains(driver)
        actions.move_to_element(target)
        actions.click(target)
        actions.perform()
        #switch tab
        driver.switch_to.window(driver.window_handles[1])
        time.sleep(7)
        driver.page_source
        #Ambil tabel tanaman
        html=driver.page_source
        soup=BeautifulSoup(html,'html.parser')
        soup
        div=soup.find_all( "table" , class_ = "org" )
        table=pd.read_html(str(div))
        hasil = table[0]
        hasil
        #simpan tabelnya
        table=pd.read_html(str(div))
        hasil = table[0]
        hasil['Compounds']=compounds
        print(hasil)
        k=k+1
        plants_df=pd.concat([plants_df,hasil], ignore_index=True)
        driver.close()
        print("restarting...")
        l=l+1
    except:
        print("No Plants found")
        k=k+1
        m=m+1
        continue
print("Total Senyawa with plants:", l)
print("Total Senyawa no plants:", m)        
plants_df
        
```

```python
plants_df.to_csv('D:\Thesis\Data\Daylight Predict Result Plants.csv', index=False)
```
