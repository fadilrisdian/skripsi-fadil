# Check Missing Data

## Train Data

```python
# Train data w Pubchem ID
train_df = pd.read_csv('../data/com-train-pos.csv')
train_df = train_df[['uniprot_id', 'drug', 'pubchemID', 'smiles']]
print(train_df.shape)
train_df.head()
```

```python
# Train data fp
train_fp_df = pd.read_csv('../data/com-train-pos-fp.csv')
train_fp_df.columns = ['pubchemID'] + list(train_fp_df.columns[1:])
print(train_fp_df.shape)
train_fp_df.head()
```

# Combine The Existing with The Missing

## Test Data

```python
test_combine = test_combine.drop(['pubchem_id'], axis=1)
test_combine.head()
```

```python
missing_test = pd.read_csv('../data/missing-test.csv')
missing_test_fp = pd.read_csv('../data/missing-test-fp.csv')
missing_test_combine = missing_test.merge(missing_test_fp, left_on='new_pubchem_id', right_on='pubchemID')
missing_test_combine = missing_test_combine.drop(['pubchem_id', 'new_pubchem_id'], axis=1)
missing_test_combine
```

```python
test_final = pd.concat([test_combine, missing_test_combine])
print(test_final.shape)
test_final.head()
```

```python
test_final.to_csv('../data/com-test-herbaldb-final.csv', index=False)
```
