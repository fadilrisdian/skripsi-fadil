```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
fp = pd.read_csv('../data/com-train-fp.csv')
fp.columns = ['drug', 'pubchemID'] + ['fp_' + str(i) for i in range(881)]
print(fp.shape)
fp.head()
```

```python
interactions = pd.read_csv('../data/train-interactions.csv')
print(interactions.shape)
interactions.head()
```

# FP-AAC

```python
aac = pd.read_csv('../data/prot-aac.csv')
aac.head()
```

```python
fp_aac = interactions.merge(aac, on='uniprot_id')
print(fp_aac.shape)
fp_aac.head()
```

```python
fp_aac = fp_aac.merge(fp, on='pubchemID')
fp_aac = fp_aac.drop('Unnamed: 0', axis=1)
print(fp_aac.shape)
fp_aac.head()
```

```python
fp_aac.to_csv('../data/train-fp-aac.csv', index=False)
```
